# Analysis of Football Match Event Data - Which players interact most?


The goal of this task is, to identify which players have the most interactions with each other and which players can be described as the center or hub of the team, meaning which player are involved in most passes. The output will be a directed network graph for each team. A directed edge from player A to player B means that player A successfully passed the ball to player B. The more passes occur between two players the wider the edge will be.
How can this be usefull for a coach?
The hypothesis is that the more successful passes occur between two players, the stronger the recommendation to have them play together in a game.

## Plan

1. Get an understanding of the data
2. Clean the data
3. Extract passing information in order to get following structure in a dataframe:
    PassEventId | TimeOfPass  | Team | PassFromPlayer | PassToPlayer|
4. Network Visualization


## Data Understanding

The data source of this task is a xml file containing event data from the game Genoa against Internationale from season 2013/2014. The complete match can be found online https://www.youtube.com/watch?v=Ea7jtP2n5PE .
Below is a screenshot of how the data looks like. Every event has certain features. The details of what each feature means can be found on Ilias: https://elearning.hslu.ch/ilias/goto.php?target=file_4821583_download&client_id=hslu . The features that are important for this task are type_id = 1, which means that the event is a pass. The attribute outcome = 1, shows that the pass was successful (0 would mean unsuccessful). Player_id and team_id give a unique number to each player and both teams.
When a player passes the ball successfully, the following player is the teammate that received the ball.

The dataframe is helpful to check if data extraction worked well. We now want to find pass-events and the two players involved in the pass. In order to do that we first need to find a pass-event with a positive outcome, the player in the following event must be the player that received the ball. First we will do two ckecks to see if the data was correctly understood:

1. Are there pass events with positive outcome where the team id of the following event is not the same as for the pass event? This should not be the case.
2. Are there pass events with positive outcome where in the next event the player id has not changed? This should not be the case.

In order to do the analysis we will us the event_list variable where each row has been saved as a tuple. This will allow us to iterate through the rows.

```python
#Check 1
#For the anomalies -if they exist - we want to know what the event after the next_event is to get a better picture of what happened.
#For every event of those anomalies we need the next_event and the after_next_event


anomalies_check1 = []

#Check the event pairs, where first event is a successful pass
for event in event_list:
    try:
        next_event = event_list[event_list.index(event)+1]
    except IndexError:
        next_event = "N/A"
    try:
        after_next_event = event_list[event_list.index(event)+2]
    except IndexError:
        after_next_event = "N/A"
    #First event should be a pass and outcome should be positive
    if event[2] == "1" and event[5] == "1":
        #Check if team of first event and next event is not same
        if event[3] != next_event[3]:
            #Append event pair to list
            anomalies_check1.append([event, next_event,after_next_event])

print(anomalies_check1)            
            
#For the anomalies we want to see of what type the "next_event" is to see why the team differs
#from the first event
anomalytypes = []

#Get the event types and append to list
for anomaly in anomalies_check1:
    anomalytypes.append(anomaly[1][2])

#Count occurences per type
Counter(anomalytypes)



```

In 68 cases when a successful pass was made by a player, the following event is not from the same team. This is a contradiction to the positive outcome of the pass. It is expected that after a successful pass, an event of a player of the same team will occure (passes the ball, gets fouled, shoots a goal etc.). In the following we will investigate further why expectations are not met.


```python
#For the diffeten event types of the "next_events" that are part of the anomalies, we want to be able to get
#all the affected datapoints
def anomaly_triplets_by_type(anomalies_list, type_of_next_event):
    for anomaly in anomalies_list:
        if anomaly[1][2] == type_of_next_event:
            print(anomaly)
            
#Insert type of "next_event" to see the triplets
anomaly_triplets_by_type(anomalies_check1, "55")

#---------------------
#In the anomalies type 44 comes up 14 times as an event after a pass.
#Type 44 is defined as "Aerial duel  50/50 when the ball is in the air 
#outcome will represent whether the duel was won or lost"

#Aerial duel (44) ---> although player of same team wins the duel, player of oppenent team comes first
#---> skip opponent player and take event after next_event

#---------------------        
#In the anomalies type 7 comes up 13 times as an event after a pass.
#"Tackle (7) = dispossesses an opponent of the ball -
#Outcome 1 = win & retain possession or out of
#play, 0 = win tackle but not possession"

#Tackle without possession of opponent ---> unsuccessful tackle of opponent comes first
#---> skip opponent player and take player id from
#event after next_event

#---------------------
#Pass (1): In the anomalies type 1 comes up 7 times as an event after a pass.
#"Any pass attempted from one player to another -
#free kicks, corners, throw-ins, goal kicks and
#goal assists"

#Failed Pass:  In some cases an unsuccessful pass was made by the opponent at the same time, in other cases
#there seem to be mistakes in the data
#---> skip opponent player and take player id from event after next_event

#---------------------
#Challenge (45): When a player fails to win the ball as an
#opponent successfully dribbles past them
#---> skip opponent player and take player id from event after next_event

#---------------------
#Deleted Events (43)
#---> skip opponent player and take player id from event after next_event

#---------------------
#Offside provoked (55)
#Awarded to last defender when an offside
#decision is given against an attacker
#Although pass was successful an opponent player caused an offside
#---> skip opponent player and take player id from event after next_event

#---------------------
#Blocked Pass (74) : Similar to interception but player already very
#close to ball
#The ball was successfully received by the player of the same team but the receiver himself tried himself to
#pass as well but was blocked. The block is listed first after the original pass.
#---> skip opponent player and take player id from event after next_event

#---------------------
#Save (10): Goalkeeper event; saving a shot on goal. Can
#also be an outfield player event with qualifier 94
#for blocked shot.
#Player that received the ball had a shot on goal. Block of goalkeeper of opponent listed first after the original pass.
#---> skip opponent player and take player id from event after next_event

#---------------------
#Clearance (12) Player under pressure hits ball clear of the
#defensive zone or/and out of play
#Player that received the ball from original pass looses the ball afterwards. Loss of ball listed first.
#---> skip opponent player and take player id from event after next_event

#---------------------
#Ball recovery (49) When a player takes possession of a loose ball
#Player that received the ball from original pass looses the ball afterwards. Loss of ball listed first.
#---> skip opponent player and take player id from event after next_event

#---------------------
#Ball touch (61) Used when a player makes a bad touch on the
#ball and loses possession. Outcome 1  ball
#simply hit the player unintentionally. Outcome 0
# Player unsuccessfully controlled the ball.
#During the transfer of the ball, opponent touches the ball
#---> skip opponent player and take player id from event after next_event

#---------------------
#Foul (4) This event ID shown when a foul is committed
#resulting in a free kick
#Opponent commits a foul. Foul listed first.
#---> skip opponent player and take player id from event after next_event

#---------------------
#Keeper Sweeper (59) When keeper comes off his line and/or out of his
#box to clear the ball
#---> skip opponent player and take player id from event after next_event

#---------------------
#Interception (8) When a player intercepts any pass event
#between opposition players and prevents the
#ball reaching its target. Cannot be a clearance.
#The original pass reaches target, receiver of the ball then tries to pass again but gets intercepted.
#The interception is listed before the second pass.
#---> skip opponent player and take player id from event after next_event



        
```

In summary the result of the investigation is that following happend:

**Instead of this order:**

Pass Player 1 of Team A at time (t-1) -->  Action Player 2 of Team A at time (t) --> Action Player 3 of Team B at time (t)


**This order was registered:**

Pass Player 1 of Team A at time (t-1) -->  Action Player 3 of Team B at time (t) --> Action Player 2 of Team A at time (t)


**Solution:**

For all the anomalies we will take the "after_next_event" as the event following the pass in order to get the correct player id.

Pass Player 1 of Team A at time (t-1) ~~-->  Action Player 3 of Team B at time (t)~~~ --> Action Player 2 of Team A at time (t)

```python
#Check 2
anomalies_check2 = []
               
for event in event_list:
    try:
        next_event = event_list[event_list.index(event)+1]
    except IndexError:
        next_event = "N/A"
    try:
        after_next_event = event_list[event_list.index(event)+2]
    except IndexError:
        after_next_event = "N/A"
    if event_list.index(event) <= len(event_list)-2:
        if event[2] == "1" and event[5] == "1":
            if event[4] == next_event[4]:
                anomalies_check2.append([event, next_event,after_next_event])

for anomaly in anomalies_check2:
    print(anomaly)

```

After the analysis of the anomalies we need to correctly modify the data in order to get to correct "from" "to" passing passing pairs.

```python
#Goal: Create dataframe with following structure:
#
#-------------------------------------------------------------------
# PassEventId | TimeOfPass  | Team | PassFromPlayer | PassToPlayer|
#-------------------------------------------------------------------
# 
#

pass_events = {}

#If the pass-event is part of the anomalies, take the "after_next_event" as the "next_event" to get the correct
#PassToPlayer in the next step
#From Check1
for event in anomalies_check1:
    #add the event and its after_next_event to the dictionary
    e_anom = (event[0] + event[2])
    pass_events[event[0][0]] = e_anom
#From Check2
pass_events[anomalies_check2[0][0][0]] = (anomalies_check2[0][0] + anomalies_check2[0][2])


#For the rest take the next_event to get the correct PassToPlayer in the next step
for event in event_list:
    try:
        next_event = event_list[event_list.index(event)+1]
    except IndexError:
        next_event = "N/A"
    if event[2] == "1" and event[5] == "1":
        #If the event id of the pass does not already exist in the dictionary
        #add the event and its next_event to the dictionary
        if event[0] not in pass_events:
            pass_events[event[0]] = (event + next_event)
print(len(pass_events))
                
#Create Dataframe
df_passes_first = pd.DataFrame.from_dict(pass_events, orient='index')


df_passes_second = df_passes_first.iloc[:,[0,1,3,4,9,10]].rename(columns={0 :"PassEventId", 1: "TimeOfPass", 3: "FromTeam", 4:"PassFromPlayer",9:"ToTeam",10:"PassToPlayer"})
#print(df_passes_second)

#Check that "FromTeam" = "ToTeam"
np.unique(df_passes_second["FromTeam"]==df_passes_second["ToTeam"])
#Yields True

df_passes_third = df_passes_second.loc[:,["PassEventId","TimeOfPass","FromTeam","PassFromPlayer","PassToPlayer"]].rename(columns={"FromTeam":"Team"})
#print(len(df_passes_third))
#yields 696

df_passes_third = df_passes_third.dropna()
#print(len(df_passes_third))
#yields 694
      
#Two rows contain None value, because there are only two they will be removed
#print(len(df_passes_third))

print(df_passes_third)
```

Let us now proceed with the visualization of the networks and see if we can find key players from a network perspective.

```python
#First let us create a simple directed graph which does not contain the weights, meaning if player A and B pass
#the ball 1 time or 50 timee, the edge will look the same.
#We will start with Internazionale and will then replicate the code for Genoa

#From df_passes_inter create a directed graph
G_Inter =  nx.from_pandas_edgelist(df_passes_inter, source='PassFromPlayer', target='PassToPlayer',create_using=nx.DiGraph())
plt.figure(3,figsize=(12,12)) 
nx.draw(G_Inter, with_labels=True, pos=nx.spring_layout(G_Inter), node_size = 500, arrows = True,arrowsize= 15,font_size = 15, edge_color = "grey")
```

The first network gives us a first insight to the data. Players that are not part of many passes are located on the outter parts of the network. Campagnaro, who we described as the key player in the first analysis does not differ too much from the other playetr. By creating and adding the weights, the differences of the player should be more visible. To create the weights we will need to count the occurences of every pass event.

```python
#Add all the pass pairs to tuples
passevents_inter = [tuple(passevent_inter) for passevent_inter in df_passes_inter.loc[:,["PassFromPlayer","PassToPlayer"]].to_numpy()]

#Create a dictionary that contains pass pair ads key and the number of occurences as value
pass_int_dict = {}
for p in set(passevents_inter):
    pass_int_dict[p] = 0

for p in passevents_inter:
    pass_int_dict[p] += 1

print(pass_int_dict)
```

```python
#Add the weights to the graph
for p in pass_int_dict:
    G_Inter[p[0]][p[1]]["weight"] = pass_int_dict[p]

#Check if weights have been added
#G_Inter.get_edge_data('F. Guarn', 'Juan Jesus')
```

```python
#Now we will draw the weighted graph


plt.figure(3,figsize=(20,20))
#pos = nx.spring_layout(G_Inter, seed=0, k=1000, iterations=50)

pos = nx.spring_layout(G_Inter, seed=0)
nx.draw_networkx(G_Inter, pos,edge_color ="white")

for edge in G_Inter.edges(data='weight'):
    nx.draw_networkx_edges(G_Inter, pos, edgelist=[edge], width=edge[2]/2, edge_color ="grey", node_size = 500, arrows = True,arrowsize= 15,font_size = 15)
    
    
    
#nx.draw_networkx_edges(G_Inter, pos = coords, edgelist=[edge], width=edge[2])




```

Compared to first network graph this one gives us more information, we can now clearly see wider edges for the players we assumed to be involved in more passes. Campagnaro does not differ out too much but it is clearly visible that he is involved in more passes than the average. The traffic between him and Jonathan seems to be the strongest.

Before repeating the analysis for Genoa, let us now focus on different metrics that gives us more insight to the network.

There are multiple ways to calculate the centrality in the network. We will calculate mutliple centrality values and add them together in a dataframe.


**Degree Centrality**

The degree centrality is defined by how many edges a node has. The more edges, incoming or outcoming, the higher the centrality. (Dr. Isabel Raabe 2020, Analysis of social Interactions).
In our case this means, the more passes a player makes or receives the more central he is. 

```python
degree_centrality_inter = nx.degree_centrality(G_Inter)
dict(sorted(degree_centrality_inter.items(), key=lambda item: item[1], reverse=True))
```

**Eigenvector Centrality**

"Eigenvector centrality assigns relative scores to all nodes in the network based on the concept that connections to high-scoring nodes contribute more to the score of the node in question than equal connections to low-scoring nodes." (https://www.geeksforgeeks.org/eigenvector-centrality-centrality-measure/ 2021)
In our case this means that players are more central when they interact with player that have themselves many interactions with many players.

```python
eigenvector_centrality_inter = nx.eigenvector_centrality(G_Inter, max_iter=100, tol=1e-06, nstart=None, weight="weight")
dict(sorted(eigenvector_centrality_inter.items(), key=lambda item: item[1], reverse=True))
```

**Closeness Centrality**

Closeness centrality checks for each node what the shortest path to all other nodes is.
The smaller the sum of those shortest paths, the more central the node is. (Dr. Isabel Raabe 2020, Analysis of social Interactions). In our case the shortest path means how many passes are needed the get the ball from player A to all other players.

```python
closeness_centrality_inter = nx.closeness_centrality(G_Inter, u=None, distance=None, wf_improved=True)
dict(sorted(closeness_centrality_inter.items(), key=lambda item: item[1], reverse=True))
```

**Betweenness Centrality**

The betweenness centrality measures for every node how many shortest paths go through it. (Dr. Isabel Raabe 2020, Analysis of social Interactions)
For every player combination, we check the shortest path in the network so that ball gets from one player to the other.

```python
#Betweenness Centrality
betweenness_centrality_inter = nx.betweenness_centrality(G_Inter, k=None, normalized=True, weight=None, endpoints=False, seed=None)
dict(sorted(betweenness_centrality_inter.items(), key=lambda item: item[1], reverse=True))
```
