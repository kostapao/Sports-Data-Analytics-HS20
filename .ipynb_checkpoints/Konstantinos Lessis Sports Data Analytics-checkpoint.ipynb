# Analysis of Football Match Event Data - Which players interact most?


The goal of this task is, to identify which players have the most interactions with each other. The output will be a directed network graph for each team. A directed edge from player A to player B means that player A successfully passed the ball to player B. The more passes occur between two players the wider the edge will be.
How can this be usefull for a coach?
The hypothesis is that the more successfull passes occur between two players, the stronger the recommendation to have them play together in a game.

## Plan

1. Get an understanding of the data
2. Turn XML data to pandas dataframe
3. Extract passing information in order to get following structure:
    Team - FromPlayer - Toplayer
4. Create Graphs in order to visualize the result


## Data Understanding

The data source of this task is a xml file continaing event data from the game Genoa against Internationale frim season 2013/2014. Below is a screenshot of how the data looks like. Every event has certain features. The details of what every feature means can be found on Ilias: https://elearning.hslu.ch/ilias/goto.php?target=file_4821583_download&client_id=hslu . The features that are important for this task are type_id = 1, which means that the event is a pass. The attribute outcome = 1, shows that the pass was successful (0 would mean unsuccessful). Player_id and team_id give a unique number to each player and both teams.
When a player playes a pass successfully, the following player is the teammate that received the ball.
